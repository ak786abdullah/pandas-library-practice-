# üîó  Concatenation with Hierarchical Indexing (MultiIndex) using keys=

When you combine multiple DataFrames using pd.concat(), you can use the keys= parameter to:

üß© Group the original DataFrames under separate labels (like folders inside a bigger folder).

This creates a MultiIndex, which helps keep track of where each row originally came from.

üß∫ Analogy:

Imagine you have two notebooks:

üìò Boys with 2 students

üìï Girls with 2 students

You stack them together in single notebook, but want to remember which student came from which notebook.
So you label   with "Boys" or "Girls" ‚Üí That‚Äôs exactly what keys= does.

```python
import pandas as pd

# üßë Boys DataFrame
boys = pd.DataFrame({
    "Name": ["Ali", "Usman", "Bilal"],
    "Math": [90, 85, 88],
    "English": [80, 78, 82]
})

# üëß Girls DataFrame
girls = pd.DataFrame({
    "Name": ["Sara", "Hina", "Zara"],
    "Math": [95, 88, 91],
    "English": [89, 92, 90]
})
```

```python
# Combine using keys
combined_notebook=pd.concat([boys,girls],axis=0,keys=["Boys","Girls"])
print(combined_notebook)
```

| Setting            | `keys` label appears on |
| ------------------ | ----------------------- |
| `axis=0` (default) | Rows                    |
| `axis=1`           | Columns                 |

#  üß† Day 23: Merging DataFrames


üìå Why Learn merge()?


You use it when you want to combine two DataFrames based on a common column, like:

Student IDs

Names

Roll numbers

merge() means connecting two DataFrames using a common piece (column)

üóÇÔ∏è DataFrame 1: Students Info

```python
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["Ali", "Sara", "Ahmed"],
    "Roll_No": [101, 102, 103],
    "Class": ["6A", "6A", "6B"]
})
print(df1)
```

üóÇÔ∏è DataFrame 2: Exam Marks

```python
df2 = pd.DataFrame({
    "Roll_No": [101, 102, 103],
    "Class": ["6A", "6A", "6B"],
    "Math": [88, 92, 75],
    "English": [90, 95, 80]
})
print(df2)
```

as we can see "Roll_No" and 'Class' both are common columns in both dataframes so for connecting them  we use merge() on Roll_No and Class

```python
merged=pd.merge(df1,df2,on=["Roll_No","Class"])
print(merged)
```

When we use pd.merge(df1, df2) without specifying on=, it will:

üîé Automatically merge on all columns that have the same name in both DataFrames.

üö® Important Note:

This only works safely when you're sure the same-named columns in both DataFrames should be used to match.
Otherwise, you should manually specify on=,

üß† What is how=?

it is perameter of merge()

defination:

,Which rows to keep when combining two DataFrames.



üìÑ df1 ‚Äî Students

```python
df1 = pd.DataFrame({
    "Roll_No": [101, 102, 103],
    "Name": ["Ali", "Sara", "Ahmed"]
})
df1
```

üìÑ df2 ‚Äî Marks

```python
df2 = pd.DataFrame({
    "Roll_No": [101, 102, 104],
    "Math": [88, 92, 79]
})
df2
```

how="inner"

```python
pd.merge(df1,df2,on="Roll_No", how="inner") #inner only keeps rows that match in both ‚Üí 101 & 102
```

 how= "left"

```python
pd.merge(df1,df2,on= "Roll_No",how="left") #left keeps all rows from df1, fills unmatched with NaN
```

how= "outer"

```python
pd.merge(df1,df2,on="Roll_No",how="outer") # outer keep all raws,fil unmatchedl with NaN)
```

how="right"

```python
pd.merge(df1,df2,on="Roll_No",how="right") #right keeps all rows from df2, fills unmatched with NaN
```

üß© Mini Project: School Result Dashboard (Using merge)
üéì Scenario:

You are a data analyst for a school. You have:

A student list

A marksheet for different subjects

A classroom assignment list

Your job is to:

Combine the data using merge()

Calculate total marks

Identify top scorers in each class

**DataFrames**

```python
import pandas as pd

# Students list
students = pd.DataFrame({
    "Roll_No": [101, 102, 103, 104],
    "Name": ["Ali", "Sara", "Ahmed", "Hina"]
})

# Marks sheet
marks = pd.DataFrame({
    "Roll_No": [101, 102, 103],
    "Math": [88, 92, 75],
    "English": [90, 95, 80]
})

# Class assignment
classes = pd.DataFrame({
    "Roll_No": [101, 102, 103, 104],
    "Class": ["6A", "6A", "6B", "6C"]
})
```

first cimbine students and marks on roll_No

```python
students_marks=pd.merge(students,marks,on="Roll_No",how="left")
students_marks
```

now combine  students_marks and classes on roll_no

Note:

i combined on "Roll_No" becouse it is common in all dataframes )

```python
students_marks_classes=pd.merge(students_marks,classes,on="Roll_No")
students_marks_classes
```

now fill NaN values with 0

```python
students_marks_classes.fillna(0,inplace=True)
students_marks_classes
```

** Add  Total_Marks**

```python
students_marks_classes["Total_Marks"]=students_marks_classes["Math"]+students_marks_classes["English"]
students_marks_classes
```

Identify top scorers in each class

```python
# Group by class and find the top scorer in each group using nlargest
top_scorers_per_class = students_marks_classes.groupby("Class", group_keys=False).apply(lambda x: x.nlargest(1, "Total_Marks"))

# Display the top scorers
print(top_scorers_per_class)
```

** now Export to CSV**

```python
students_marks_classes.to_csv("students_result.csv",index=False)
pd.read_csv("students_result.csv")
```

# Time Series Analysis ‚Äì "Working with Dates"

**goal:** Learn to handle,manipulate and analyze data that is recorded over time  

so far, we treated date as string (e.g "2025-11-10").
We needs to tell pandas that these are actual dates

**Converting to Datetime:**

The most crucial first step is converting date columns from plain text (string or object) into a special Pandas **datetime** type.
> pd.to_datetime(date_column)

Most of the time,  dates are in a standard format that pandas already recognizes (like YYYY-MM-DD or MM/DD/YYYY). In this case, **pd.to_datetime()** just works it automatically. don't need to give  formet parameter





```python
import pandas as pd
# Sample sales data with dates as strings
sales_data = pd.DataFrame({
    'sale_date': ['2025-11-01', '2025-11-02', '2025-11-03', '2025-11-04', '2025-11-05'],
    'items_sold': [25, 30, 22, 28, 35]
})
print(f"sales_data:{sales_data}\n Data_Type:\n {sales_data.dtypes}")
```

Here data type of "sales_date" is object(string),need to convert it into datetime type

```python
sales_data["sale_date"]=pd.to_datetime(sales_data["sale_date"])
print(f"Data_Type:\n {sales_data.dtypes}")
```

 The Dtype for sale_date changed from object(string) to datetime64[ns]. Now Pandas understands time!



**Handling Non-Standard Formats:**

If data given
 **non-standard format( DD-MM-YYYY):**  10-01-2024  Is January 10th

Pandas always recognize standard format(MM-DD-YYYY or YYYY-MM-DD) by default

It consider 10-01-2024 is
 October 1st (a MM-DD-YYYY standard format )

**The Solution: We provide the format.**

%d = day

%m = month

%Y = 4-digit year

%y=2-digit year

saperate %y,%d,%m by given saperator

> pd.to_datetime(date_column,format)




let we have dates from a European system (Day-Month-Year).

```python
dates_list = ['10-01-2024', '12-01-2024', '15-01-2024'] # All are January (DD-MM-YYYY ) Here saperator="-"
pd.to_datetime(dates_list,format="%d-%m-%Y")
```

**Practice question:** You have a list of dates in strings: ['31/12/2023', '01/01/2024']. How would you convert this into datetime and the format parameter? (Hint: The separator is /.)

```python
date_string= ['31/12/2023', '01/01/2024']
pd.to_datetime(date_string,format="%d/%m/%Y")
```

**Handling Bad Data:**

if date_column contain the some data that are not dates  then pd.to_datetime() raise an error

To handle this satuation we us **errors** parameter

> pd.to_datetime(date_column,errors)

**errors='raise' (The default):** Stops and raises an error.

**errors='coerce' (The flexible one):** If it can't convert a date, it will set it to NaT (Not a Time), which is the datetime version of NaN (Not a Number).



```python
dirty_dates = ['2024-01-01', '2024-01-02', 'this is not a date', '2024-01-04']
pd.to_datetime(dirty_dates,errors="coerce")
```

It just marked the bad one as NaT. This is extremely useful for cleaning messy datasets.

**Arthimatic operation in datetime:**




> **pd.Timestamp():** Means Full date (e.g 2023-02-28 00:00:00)



>**pd.Timedelta():** Means fixed duration of time.
Allowed arguments are [weeks, days, hours, minutes, seconds, milliseconds, microseconds, nanoseconds]

Notice that Months and Years are NOT in that list.

Because a "month" is not a fixed unit of time. Is it 28 days? 30 days? 31 days?

```python
# always end_time > start_time
data = {
    'Project_Start': ['2023-01-01', '2023-02-10', '2023-04-05'],
    'Project_End': ['2023-01-15', '2023-02-20', '2023-04-10']
}
df = pd.DataFrame(data)

# Step 1: ALWAYS convert to datetime first!
df["Project_Start"]=pd.to_datetime(df["Project_Start"])
df["Project_End"]=pd.to_datetime(df["Project_End"])
# Step 2: The Arithmetic
df["Project_End"]-df["Project_Start"] # it give a Timedelta(duration)
# add 5 days in Project_End
df["Project_End"]+pd.Timedelta(days=5)
# add 2 weeks in project_end
df["Project_End"]+pd.Timedelta(weeks=2)
# add 12 hours , 20 mintues and 30 second in project_end
df["Project_End"]+pd.Timedelta(hours=12,minutes=20,seconds=30)


# subtract 5 days from Project_End
df["Project_End"]-pd.Timedelta(days=5)
# subtract 2 weeks from project_end
df["Project_End"]-pd.Timedelta(weeks=2)
# subtract 12 hours , 20 mintues and 30 second from project_end
df["Project_End"]-pd.Timedelta(hours=12,minutes=20,seconds=30)
```

**we know that units must be constant. But in the Gregorian Calendar, the unit "Month" is variable.**

January = 31 days

February = 28 (or 29) days

April = 30 days   

**Timedelta** is for constant(fixed) duration (e.g 1day=24 hours are fixed duration of 1 day)

**The Problem:** A customer subcribe a monthly internet package on January 31st. we want to End this package  exactly one month later.

```python
import pandas as pd
start_date = pd.Timestamp('2023-01-31')

# ATTEMPT 1: Using Timedelta (The "Dumb" Way)
# We assume a month is 30 days.
wrong_date=start_date+pd.Timedelta(days=30)
print(wrong_date)# output:2023-03-02 00:00:00
# WAIT! We skipped February entirely! The customer gets 2 free days.
# If we used 28 days, we'd be wrong in leap years.
```

**Now Question is  how to solve this problem**

**The Solution: pd.DateOffset:** Dataoffset is not just for  a fixed duration. it is set of instructions that tell python: "Look at the calendar, figure out the logic, and move forward."

**Syntax:** pd.DateOffset(months=X, years=Y, days=Z)

Allowed arguments are [years,months,weeks, days, hours, minutes, seconds, milliseconds, microseconds, nanoseconds]  **it is advanced version of Timedelta**


```python
# ATTEMPT 2: Using DateOffset (The "Smart" Way)
correct_date=start_date+pd.DateOffset(months=1)
print(correct_date) # 2023-02-28 00:00:00
#
```

**Rules And Summary**

**pd.Timestamp:** Represents a specific point in time (e.g., '2023-01-01').

**pd.Timedelta:** Represents a fixed duration of time (e.g., '5 days', '3 hours').

**pd.DateOffset:** Represents a flexible calendar-aware duration (e.g., '1 month', which adjusts for month lengths).

**Arithmetic Rules:**

**1. Operations involving pd.Timestamp (a specific point in time):**

**Timestamp + Timedelta:** Returns a new Timestamp. (e.g., Jan 1 + 5 days = Jan 6)

**Timestamp + DateOffset:** Returns a new Timestamp. (e.g., Jan 31 + 1 month = Feb 28/29)

**Timestamp - Timedelta:** Returns a new Timestamp.

**Timestamp - DateOffset:** Returns a new Timestamp.

**Timestamp / or * Timedelta:** not allowed

**Timestamp / or * DateOffset:** not allowed

**Timestamp - Timestamp:** Returns a Timedelta (the duration between two points).

**Timestamp + or * or / Timestamp:** Not allowed (adding two specific dates doesn't make logical sense).

**Timestamp * Scalar (number)**:  Not allowed.

**Timestamp / Scalar (number):** Not allowed.

**2. Operations involving pd.Timedelta (a fixed duration):**

**Timedelta + or -  Timedelta:** Returns a new Timedelta.

**Timedelta * Timedelta:** not allowed

**Timedelta  * or / Scalar (number):** Returns a new Timedelta.

**Timedelta  + or - Scalar** not allowed

**Timedelta / Timedelta:** Returns a Scalar (a ratio, e.g., 10 days / 5 days = 2.0).

**Timedelta * Timedelta:** Not allowed.

**3. Operations involving pd.DateOffset (a flexible calendar-aware duration):**

**DateOffset + or - DateOffset:** Returns a new DateOffset. (e.g., pd.DateOffset(days=30) + pd.DateOffset(days=1) should give pd.DateOffset(days=31))

**Note:** In your current environment, direct addition of two DateOffset objects may raise a TypeError, requiring manual summation of components as we saw with the workaround.
DateOffset - DateOffset: Returns a new DateOffset.

**DateOffset + or - Scalar (number):** not allowed

**DateOffset * Scalar (number):** Returns a new DateOffset.

**DateOffset / Scalar (number):** Not directly allowed to return a DateOffset; often results in implicit conversion or error if the offset is not easily divisible.

**DateOffset / DateOffset:** Not allowed (due to their flexible nature, division between them is ambiguous).
