üìÖ Day 12: Handling Missing Values
üß† Goal:

Deal with empty/missing cells.

```python
df.isnull().sum() # checking missing value
df.dropna(inplace=True)# delete all raws [which has any missing value ]
df.dropna(how="all",inplace=True) # delete all raws [which has all values are missing]
df.dropna(subset=["Total_marks"]) # delete all raws [if any value missing in "Total_marks"]
df.fillna(0,inplace=True)# fill all missing value with zero
df.fillna(df.mode(),inplace=True) # fill all missing value with mode of df
df.bfill(inplace=True) # backword fill
df.ffill(inplace=True)# forword fill
# I use "inplace=True" means i apply these function on my original dataframe
```

üìÖ Day 17: Grouping with groupby() üß† Goal: Group data and calculate things like total, average, etc.

```python
data = {
    'Team': ['Red', 'Blue', 'Red', 'Blue', 'Red', 'Green', 'Blue', 'Green'],
    'Player': ['Ali', 'Sara', 'Hassan', 'Ayesha', 'Omar', 'Bilal', 'Zara', 'Nadia'],
    'Score': [10, 20, 15, 30, 25, 40, 35, 50]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["Total_score"]=df.groupby("Team")["Score"].transform("sum")
df
```

```python
df.groupby("Team")["Score"].mean()
```

 first it  make three groups (Blue, Green, Red) in "teams " column (combine the same values ) then calculate mean of score of each group

```python
df.groupby("Team")["Player"].count() # number of players in each team
```

üìÖ Day 18: Multiple Aggregations
üß† Goal:

Get multiple summaries of multiple columns  at once using agg().

```python
df.agg({"Score":["mean","median"]}) # mean and median of score
df.groupby("Team")["Score"].agg(["mean","median"])
#or
df.groupby("Team").agg({"Score":["mean","median"]}) # mean and median of score in each team
df.groupby("Team").agg({"Score":["mean","median"],"Total_score":"mean"}) # mean and median of score and mean of total score of each team
```

üìÖ Day 19: Sorting After Grouping
üß† Goal:

See top groups ‚Äî e.g., which team has highest average

```python
df.groupby('Team')["Score"].mean().sort_values(ascending=False)
```
